# f(x) = sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)
# [1, 30]

# Задание 1

```python
from math import sin, exp
from scipy.linalg import solve
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize, differential_evolution

def f(x):
    return sin(x/5.0)*exp(x/10.0) + 5*exp(-x/2.0)
```

```python
x1 = 1.0
x2 = 30.0

x_list = []
y_list = []
n = np.arange(1.0, 30.0, 0.1);
for x in n:
    x_list.append(x)
    y_list.append(f(x))

plt.plot(x_list, y_list)
plt.show()
```

```python
initial_guess = [2.0]
result = minimize(f, initial_guess, method='BFGS')
print(result)
```

```python
initial_guess = [30.0]
result = minimize(f, initial_guess, method='BFGS')
print(round(result.fun, 2))
```

```python
print(2/3, 2.0/3.0)
```

# Задание 2

```python
bounds = [(1.0, 30.0)]
result = differential_evolution(f, bounds)

print(round(result.fun, 2))
```

ответ на задание 2: -11.90
