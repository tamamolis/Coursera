# f(x) = sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)

```python
from math import sin, exp
from scipy.linalg import solve
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return sin(x/5.0)*exp(x/10.0) + 5*exp(-x/2.0)
```

```python
x1 = 1.0
x2 = 15.0

x_list = []
y_list = []
n = np.arange(1.0, 15.0, 0.1);
for x in n:
    x_list.append(x)
    y_list.append(f(x))

plt.plot(x_list, y_list)
plt.show()
```

```python
A = np.array([[1, x1],[1, x2]])
B = np.array([f(x1), f(x2)])
X = solve(A, B)
print(X)
```

```python
def approx_f(x, X):
    return X[0] + X[1]*x
```

```python
x_list = []
y_list = []
n = np.arange(1.0, 15.0, 0.1);
for x in n:
    x_list.append(x)
    y_list.append(approx_f(x, X))

plt.plot(x_list, y_list)
plt.show()
```

```python
x1 = 1.0
x2 = 8.0
x3 = 15.0
```

```python
def approx_second_f(x, X):
    return X[0] + X[1]*x + X[2]*x**2
```

```python
A = np.array([[1, x1, x1**2],[1, x2, x2**2], [1, x3, x3**2]])
B = np.array([f(x1), f(x2), f(x3)])
X = solve(A, B)
print(X)
```

```python
x_list = []
y_list = []
n = np.arange(1.0, 15.0, 0.1);
for x in n:
    x_list.append(x)
    y_list.append(approx_second_f(x, X))

plt.plot(x_list, y_list)
plt.show()
```

```python
x1 = 1.0
x2 = 4.0
x3 = 10.0
x4 = 15.0
```

```python
def approx_third_f(x, W):
    return W[0] + W[1]*x + W[2]*x**2 + W[3]*x**3
```

```python
A = np.array([[1, x1, x1**2, x1**3],[1, x2, x2**2, x2**3], [1, x3, x3**2, x3**3], [1, x4, x4**2, x4**3]])
B = np.array([f(x1), f(x2), f(x3), f(x4)])
W = solve(A, B)
# print(W[0], W[1], W[2], W[3])
print(np.round(W, 3))
print(f(x1), f(x2), f(x3), f(x4))
```

```python
x_list = []
y_list = []
n = np.arange(1.0, 15.0, 0.1);
for x in n:
    x_list.append(x)
    y_list.append(approx_third_f(x, W))

plt.plot(x_list, y_list)
plt.show()
```
