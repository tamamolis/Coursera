```python
import numpy as np
import re
```

```python
list_of_words = []
list_of_sentences = []
number_of_lines = 0

f = file('sentences.txt', 'r')
for line in f:
    number_of_lines+=1
    string = line.lower()
    string = re.split('[^a-z]', string)
    list_of_sentences.append(string)
    number_of_lines += 1
f.close()

for sentences in list_of_sentences:
    list_of_words.append(filter(None, sentences))
```

```python
mydict = {}
d = 0 
for sentences in list_of_words:
    for item in sentences:  
        if mydict.has_key(item):  
            pass  
        else:  
            mydict[item] = d
            d+=1
```

```python
reverse_dict = {}
reverse_dict = dict(zip(mydict.values(), mydict.keys()))
```

```python
from collections import Counter

def count_word(sentences, word):
    counter = Counter(sentences)
    element = counter[word]
    return element
```

```python
d = len(reverse_dict.keys())
n = number_of_lines
array = np.zeros((n, d))
for i in range(n):
    for j in range(d):
        word = reverse_dict[j]
        element = count_word(list_of_words[i], word)
        array[i][j] = element
```

```python
from scipy.spatial.distance import cosine

input_line = array[0]
buf = 100
index = 0
for i in range(1, n):
    cosinus = cosine(input_line, array[i])
    if cosinus < buf:
        buf = cosinus
        index = i
print(buf, index)
```
