```python
import pandas as pd
import sklearn
from pandas.compat import StringIO
import csv
import numpy as np
from sklearn.cluster import MeanShift
```

```python
with open('checkins.dat') as dat_file, open('checkins.csv', 'w') as csv_file:
    csv_writer = csv.writer(csv_file)
    for line in dat_file:
        row = [field.strip() for field in line.split('|')]
        if len(row) == 6 and row[3] and row[4]:
            csv_writer.writerow(row)
```

```python
path = 'checkins.csv'
# fields = ['id','user_id', 'venue_id', 'latitude', 'longitude', 'created_at'])
fields = ['latitude', 'longitude']

df = pd.read_csv(path, 
                usecols=fields)
```

```python
df.isnull().values.any()
```

```python
df = df[0:100000]
```

```python
df.head()
```

```python
centres = [ [33.751277, -118.188740], [25.867736, -80.324116], [51.503016, -0.075479],
[52.378894, 4.885084],[39.366487, 117.036146], [33.868457, 151.205134] ]
```

```python
centres = np.array(centres)
```

```python
np.shape(centres)
```

```python
train_data = df.iloc[0:100000]
```

```python
ms = MeanShift(bandwidth=0.1)
ms.fit(train_data)
```

```python
labels = ms.labels_
cluster_centers = ms.cluster_centers_

labels_unique = np.unique(labels)
n_clusters_ = len(labels_unique)

print("number of estimated clusters : %d" % n_clusters_)
```

```python
min = 100000
point = np.zeros((1, 2))

for coord_cl in cluster_centers:
    for coord_of in centres:
        dist = np.linalg.norm(coord_of-coord_cl)
        if dist < min:
            print(dist, min, coord_cl)
            point[0][0] = coord_cl[0]
            point[0][1] = coord_cl[1]
            min = dist
```
